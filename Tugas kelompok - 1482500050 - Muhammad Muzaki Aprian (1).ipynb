```python
class User:
    def __init__(self, username):
        self.username = username

    def tampil_info(self):
        print(f"Username : {self.username}")


class Admin(User):
    def __init__(self, username):
        super().__init__(username)


class Anggota(User):
    def __init__(self, username, id_anggota):
        super().__init__(username)
        self.id_anggota = id_anggota

    def tampil_info(self):
        print(f"ID Anggota : {self.id_anggota}")
        print(f"Nama       : {self.username}")
        print("-" * 30)


class Buku:
    def __init__(self, id_buku, judul, penulis, stok):
        self.id_buku = id_buku
        self.judul = judul
        self.penulis = penulis
        self.stok = stok

    def tampil_info(self):
        print(f"ID Buku    : {self.id_buku}")
        print(f"Judul      : {self.judul}")
        print(f"Penulis    : {self.penulis}")
        print(f"Stok       : {self.stok}")
        print("-" * 30)


class Perpustakaan:
    def __init__(self):
        self.daftar_buku = []
        self.daftar_anggota = []

    def tambah_buku(self, buku):
        self.daftar_buku.append(buku)
        print("Buku berhasil ditambahkan!\n")

    def tampilkan_buku(self):
        if len(self.daftar_buku) == 0:
            print("Belum ada data buku\n")
        else:
            print("\n===== DAFTAR BUKU =====")
            for buku in self.daftar_buku:
                buku.tampil_info()

    def tambah_anggota(self, anggota):
        self.daftar_anggota.append(anggota)
        print("Anggota berhasil ditambahkan!\n")

    def tampilkan_anggota(self):
        if len(self.daftar_anggota) == 0:
            print("Belum ada data anggota\n")
        else:
            print("\n===== DAFTAR ANGGOTA =====")
            for anggota in self.daftar_anggota:
                anggota.tampil_info()


perpus = Perpustakaan()

while True:
    print("===== SISTEM MANAJEMEN PERPUSTAKAAN =====")
    print("1. Tambah Buku")
    print("2. Lihat Buku")
    print("3. Tambah Anggota")
    print("4. Lihat Anggota")
    print("5. Keluar")

    pilihan = input("Pilih Menu : ")

    if pilihan == "1":
        id_buku = input("Masukkan ID Buku : ")
        judul = input("Masukkan Judul Buku : ")
        penulis = input("Masukkan Penulis Buku : ")
        stok = int(input("Masukkan Stok Buku : "))

        buku = Buku(id_buku, judul, penulis, stok)
        perpus.tambah_buku(buku)

    elif pilihan == "2":
        perpus.tampilkan_buku()

    elif pilihan == "3":
        username = input("Masukkan Nama Anggota : ")
        id_anggota = input("Masukkan ID Anggota : ")

        anggota = Anggota(username, id_anggota)
        perpus.tambah_anggota(anggota)

    elif pilihan == "4":
        perpus.tampilkan_anggota()

    elif pilihan == "5":
        print("Program selesai")
        break

    else:
        print("Menu tidak tersedia\n")
```
